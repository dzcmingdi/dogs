```python
import win32com.client as client
import os
# pip install pywin32
# 说话
# speak = client.Dispatch("SAPI.SpVoice")
# speak.Speak("asff")
# 保存文件
def save_txt(txt_dir, save_dir):
    fileStream = client.Dispatch("SAPI.SpFileStream")
    fileStream.Open(save_dir, 3, False)
    spVoice = client.Dispatch("SAPI.SpVoice")
    spVoice.AudioOutputStream = fileStream
    with open(txt_dir, encoding='utf-8',mode="r") as f:  # 打开文件
            data = f.read()  # 读取文件
    spVoice.speak(data)
    fileStream.Close()
img_dir = 'data/dog_data'
path_all = []
for path in os.listdir(img_dir):
    path_all.append(path)
for path in path_all:
    txt_path = img_dir + '/' + path + '/' + path + '.txt'
    print(txt_path)
    save_path = img_dir + '/' + path + '/' + path + '.wav'
    print(save_path)
    save_txt(txt_path, save_path)



```
